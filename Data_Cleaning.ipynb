```python
import pandas as pd
AirQuality_world = pd.read_csv('/content/AQI and Lat Long of Countries.csv')
AirQuality_world.head()
```

```python
len(AirQuality_world)
```

```python
AirQuality_world.isnull().sum()
```

```python
Air1 = AirQuality_world

import geopy.geocoders as geocoders

locator = geocoders.Nominatim(user_agent='my_app')

for i in range(len(Air1)):

        lat = Air1.iloc[i]['lat']
        lon = Air1.iloc[i]['lng']

        if not pd.isna(Air1.loc[i, 'Country']):
            continue

        location = locator.reverse((lat, lon), exactly_one=True)

        if isinstance(location.address, str):
            country = location.address.split(', ')[-1]
        else:
            country = location.address.get('country', None)
        Air1.loc[i, 'Country'] = country

Air1 = pd.DataFrame(Air1)
Air1
```

```python
Air1.isnull().sum()
```

```python
Air1.drop_duplicates(subset=['Country','City','AQI Value','AQI Category','CO AQI Value','CO AQI Category','Ozone AQI Value','Ozone AQI Category',
                             'NO2 AQI Value','NO2 AQI Category','PM2.5 AQI Value','PM2.5 AQI Category'], inplace=True)
print(Air1)
```

```python
Air1.to_csv('AirQuality.csv')
```
